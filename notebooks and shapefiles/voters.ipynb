## Cleaning election voters data from Comelec

```python
import pandas as pd
import numpy as np
import fuzzy_pandas as fpd
pd.set_option('display.max_columns', None)
```

```python
election_2019 =pd.read_excel("ph-election-voters/2019 voters.xlsx", sheet_name='2019_municipality_final')
election_2022 =pd.read_excel("ph-election-voters/2022 voters.xlsx", sheet_name='2022_municipality_final')
```

```python
election_2022.province = election_2022.province.str.strip()
election_2022.region = election_2022.region.str.strip()
election_2022.municipality = election_2022.municipality.str.strip()
election_2022.legislative_district = election_2022.legislative_district.str.strip()
election_2022
```

```python
election_2019.province = election_2019.province.str.strip()
election_2019.region = election_2019.region.str.strip()
election_2019.municipality = election_2019.municipality.str.strip()
election_2019.municipality = election_2019.municipality.str.strip()
election_2019.legislative_district = election_2019.legislative_district.str.strip()
election_2019
```

```python
election_2019.municipality = election_2019.municipality.str.replace(r'\(.*?\)', '', regex=True)
election_2019
```

```python
election_2019.municipality = election_2019.municipality.str.strip()
```

```python
merged = pd.merge(election_2019, election_2022, on = ['region', 'province', 'municipality', 'legislative_district'], how = 'outer')
merged
```

```python
merged.to_csv('merged.csv', index=False)
```
