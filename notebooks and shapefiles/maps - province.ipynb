## Mapping voters - province

For voter turnout.

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import fuzzy_pandas as fpd
from plotnine import *
pd.set_option('display.max_columns', None)
```

## Load merged file

```python
legis_district =pd.read_excel("merged_legislative_district.xlsx", sheet_name='merged')
legis_district
```

```python
legis_district.province = legis_district.province.str.title()
```

## Group by province

```python
by_province1 = legis_district.groupby(['province']).a2022_total_voters.mean().to_frame().reset_index()
by_province2 = legis_district.groupby(['province']).a2022_total_voter_actual.mean().to_frame().reset_index()
```

## Merge all three dfs

```python
by_province_2022 = pd.merge(by_province1, by_province2, on = ['province'], how = 'outer')
by_province_2022
```

```python
by_province_2022['a2022_ave_voter_turnout'] = by_province_2022.a2022_total_voter_actual / by_province_2022.a2022_total_voters *100
by_province_2022
```

## Geopandas

```python
province = gpd.read_file('phl_admbnda_adm2_psa_namria_20231106.zip')
province
```

```python
province.columns = province.columns.str.lower()
province
```

```python
province = province.drop(['adm2_pcode','adm1_pcode', 'adm0_en', 'adm0_pcode', 'validon', 'validto', 'date', 'shape_leng', 'shape_area', 'area_sqkm'], axis=1)
province.head()
```

```python
province.adm2_en = province.adm2_en.str.replace('Metropolitan Manila Fourth District', "NCR", regex=False)
province.adm2_en = province.adm2_en.str.replace('Metropolitan Manila Third District', "NCR", regex=False)
province.adm2_en = province.adm2_en.str.replace('Metropolitan Manila Second District', "NCR", regex=False)
province.adm2_en = province.adm2_en.str.replace('Metropolitan Manila First District', "NCR", regex=False)
province.adm2_en = province.adm2_en.str.replace('(not a province)', "", regex=False)
```

```python
#put NCR cities back to "City of" format instead of "City" to preserve original data
province.adm2_en = province.adm2_en.str.replace('Manila City', 'City of Manila', regex=False)
province.adm2_en = province.adm2_en.str.replace('Malabon City', 'City of Malabon', regex=False)
province.adm2_en = province.adm2_en.str.replace('Mandaluyong City', 'City of Mandaluyong', regex=False)
province.adm2_en = province.adm2_en.str.replace('Pasig City', 'City of Pasig', regex=False)
province.adm2_en = province.adm2_en.str.replace('Caloocan City', 'City of Caloocan', regex=False)
province.adm2_en = province.adm2_en.str.replace('San Juan City', 'City of San Juan', regex=False)
province.adm2_en = province.adm2_en.str.replace('Marikina City', 'City of Marikina', regex=False)
province.adm2_en = province.adm2_en.str.replace('Navotas City', 'City of Navotas', regex=False)
province.adm2_en = province.adm2_en.str.replace('Valenzuela City', 'City of Valenzuela', regex=False)
province.adm2_en = province.adm2_en.str.replace('Taguig City', 'City of Taguig', regex=False)
province.adm2_en = province.adm2_en.str.replace('Makati City', 'City of Makati', regex=False)
province.adm2_en = province.adm2_en.str.replace('Paranaque City', 'City of Paranaque', regex=False)
province.adm2_en = province.adm2_en.str.replace('Muntinlupa City', 'City of Muntinlupa', regex=False)
province.adm2_en = province.adm2_en.str.replace('Las Pinas City', 'City of Las Pinas', regex=False)
```

```python
province.adm2_en = province.adm2_en.str.strip()
```

```python
by_province_2022.province = by_province_2022.province.str.strip()
```

```python
# legis_district[legis_district.municipality=='City of Mandaluyong']
```

```python
# municipal[municipal.adm3_en=='Manila City']
```

```python
final_df_final = fpd.fuzzy_merge(by_province_2022, province,
                        left_on=['province'],
                        right_on=['adm2_en'],
                        method='exact',
                        threshold=0.7,
                        ignore_case=True,
                        keep='all')
final_df_final
```

```python
final_df_final.a2022_ave_voter_turnout.sort_values(ascending=False)
```

## Create bins

```python
#2022 bins
final_df_final['prov_ave_turnout_2022_bins'] = pd.cut(np.array(final_df_final['a2022_ave_voter_turnout']),
       [65, 71, 76, 81, 86, 91, 96, 100], labels=["65-70", "71-75", "76-80", "81-85", "86-90", "91-95", "96-100"])
final_df_final
```

```python
# #2019 bins
# final_df_final['ave_turnout_2019_bins'] = pd.cut(np.array(final_df_final['2019_average_voter_turnout']),
#        [42, 51, 56, 61, 66, 71, 76, 81, 86, 91, 96, 100], labels=["45-50", "51-55", "56-60", "61-65", "66-70", "71-75", "76-80", "81-85", "86-90", "91-95", "96-100"])
# final_df_final
```

## Convert the bins to string

```python
final_df_final.prov_ave_turnout_2022_bins = final_df_final.prov_ave_turnout_2022_bins.astype(str)
# final_df_final.ave_turnout_2019_bins = final_df_final.ave_turnout_2019_bins.astype(str)
final_df_final
```

## Convert 'geometry' column back to a geopandas df

```python
final_df = gpd.GeoDataFrame(final_df_final, crs="EPSG:4326", geometry=final_df_final.geometry)
```

## Convert to GEOJSON

```python
final_df.to_file('provincial_voters.geojson', driver='GeoJSON')
```
