```python
import pandas as pd
import numpy as np
import geopandas as gpd
import fuzzy_pandas as fpd
from plotnine import *
pd.set_option('display.max_columns', None)
```

## Load merged file

```python
legis_district =pd.read_excel("merged_legislative_district.xlsx", sheet_name='merged')
legis_district
```

```python
voter_turnout= legis_district.filter(['region','province','municipality', 'a2019_total_voter_actual', 'a2022_total_voter_actual'], axis=1)
voter_turnout
```

```python
voter_turnout['Change'] = voter_turnout.a2022_total_voter_actual - voter_turnout.a2019_total_voter_actual
voter_turnout['Percentage Change'] = (voter_turnout.a2022_total_voter_actual - voter_turnout.a2019_total_voter_actual)/ voter_turnout.a2019_total_voter_actual * 100
voter_turnout
```

```python
voter_turnout.a2019_total_voter_actual = voter_turnout.a2019_total_voter_actual.map('{:,}'.format)
voter_turnout.a2022_total_voter_actual = voter_turnout.a2022_total_voter_actual.map('{:,}'.format)
voter_turnout.Change = voter_turnout.Change.map('{:,}'.format)
voter_turnout
```

```python
voter_turnout.columns = ['Region', 'Province', 'Municipality', 'Voter turnout (2019)', 'Voter turnout (2022)', 'Change', 'percent_change']
voter_turnout
```

```python
voter_turnout.percent_change = voter_turnout.percent_change.round(2)
voter_turnout.columns = ['Region', 'Province', 'Municipality', 'Voter turnout (2019)', 'Voter turnout (2022)', 'Change', 'Percent change']
```

```python
voter_turnout
```

```python
voter_turnout.Province = voter_turnout.Province.str.title()
voter_turnout.Province = voter_turnout.Province.str.replace('Ncr', "NCR", regex=False)
voter_turnout
```

```python
voter_turnout.to_csv('voter_turnout.csv', index=False)
```

## Geopandas

```python
municipal = gpd.read_file('phl_admbnda_adm3_psa_namria_20231106.zip')
municipal
```

```python
municipal.columns = municipal.columns.str.lower()
municipal
```

```python
municipal = municipal.drop(['adm3_pcode', 'adm2_pcode','adm1_pcode', 'adm0_en', 'adm0_pcode', 'validon', 'validto', 'date', 'adm3_ref','adm3alt1en', 'shape_leng', 'shape_area', 'area_sqkm'], axis=1)
municipal.head()
```

```python
municipal.adm3_en = municipal.adm3_en.str.replace('(Capital)', "", regex=False)
municipal.adm2_en = municipal.adm2_en.str.replace('Metropolitan Manila Fourth District', "NCR", regex=False)
municipal.adm2_en = municipal.adm2_en.str.replace('Metropolitan Manila Third District', "NCR", regex=False)
municipal.adm2_en = municipal.adm2_en.str.replace('Metropolitan Manila Second District', "NCR", regex=False)
municipal.adm2_en = municipal.adm2_en.str.replace('Metropolitan Manila First District', "NCR", regex=False)
municipal.adm2_en = municipal.adm2_en.str.replace('(not a province)', "", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('City of Masbate', "Masbate City", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('City of Valencia', "Valencia City", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('City of Malaybalay', "Malaybalay City", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Mataasnakahoy', "Mataas na Kahoy", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Shariff Saydona Mustapha', "Shariff Saydona Musthapa", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Tagoloan II', 'Tagoloan', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('ñ', "n", regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace(r'\s?\(.*\)', '', regex=True) #removes all words in parentheses
municipal.adm3_en = municipal.adm3_en.str.replace(r'^City of (.*)', r'\1 City', regex=True) #makes City of just City
municipal.adm3_en = municipal.adm3_en.str.replace(r'\s+', ' ', regex=True)
```

```python
#put NCR cities back to "City of" format instead of "City" to preserve original data
municipal.adm3_en = municipal.adm3_en.str.replace('Manila City', 'City of Manila', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Malabon City', 'City of Malabon', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Mandaluyong City', 'City of Mandaluyong', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Pasig City', 'City of Pasig', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Caloocan City', 'City of Caloocan', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('San Juan City', 'City of San Juan', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Marikina City', 'City of Marikina', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Navotas City', 'City of Navotas', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Valenzuela City', 'City of Valenzuela', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Taguig City', 'City of Taguig', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Makati City', 'City of Makati', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Paranaque City', 'City of Paranaque', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Muntinlupa City', 'City of Muntinlupa', regex=False)
municipal.adm3_en = municipal.adm3_en.str.replace('Las Pinas City', 'City of Las Pinas', regex=False)
```

```python
municipal.adm3_en = municipal.adm3_en.str.strip()
municipal.adm2_en = municipal.adm2_en.str.strip()
```

```python
legis_district.municipality = legis_district.municipality.str.strip()
legis_district.province = legis_district.province.str.strip()
```

```python
# legis_district[legis_district.municipality=='City of Mandaluyong']
```

```python
# municipal[municipal.adm3_en=='Manila City']
```

```python
final_df_final = fpd.fuzzy_merge(legis_district, municipal,
                        left_on=['province', 'municipality'],
                        right_on=['adm2_en', 'adm3_en'],
                        method='exact',
                        threshold=0.7,
                        ignore_case=True,
                        keep='all')
final_df_final
```

## Create bins

```python
#2022 bins
final_df_final['ave_turnout_2022_bins'] = pd.cut(np.array(final_df_final['a2022_average_voter_turnout']),
       [45, 51, 56, 61, 66, 71, 76, 81, 86, 91, 96, 100], labels=["45-50", "51-55", "56-60", "61-65", "66-70", "71-75", "76-80", "81-85", "86-90", "91-95", "96-100"])
final_df_final
```

```python
#2019 bins
final_df_final['ave_turnout_2019_bins'] = pd.cut(np.array(final_df_final['a2019_average_voter_turnout']),
       [42, 51, 56, 61, 66, 71, 76, 81, 86, 91, 96, 100], labels=["45-50", "51-55", "56-60", "61-65", "66-70", "71-75", "76-80", "81-85", "86-90", "91-95", "96-100"])
final_df_final
```

## Convert the bins to string

```python
final_df_final.ave_turnout_2022_bins = final_df_final.ave_turnout_2022_bins.astype(str)
final_df_final.ave_turnout_2019_bins = final_df_final.ave_turnout_2019_bins.astype(str)
final_df_final
```

## Convert 'geometry' column back to a geopandas df

```python
final_df = gpd.GeoDataFrame(final_df_final, crs="EPSG:4326", geometry=final_df_final.geometry)
```

## Convert to GEOJSON

```python
final_df.to_file('municipal_voters.geojson', driver='GeoJSON')
```
