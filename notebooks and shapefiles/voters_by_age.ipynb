## Cleaning election voters data from Comelec (age,sex)

Sample data is 2019

```python
import pandas as pd
import numpy as np
import fuzzy_pandas as fpd
from plotnine import *
pd.set_option('display.max_columns', None)
```

```python
election_2019_age =pd.read_excel("2019/2019_by_age_sex.xlsx", sheet_name='municipality_final_wide')
```

```python
election_2019_age
```

## For 2019 data, let's remove individual data for ages 60-over 65

```python
election_2019_age = election_2019_age.drop(['60_m', '60_f', '61_m', '61_f', '62_m', '62_f', '63_m', '63_f', '64_m', '64_f'], axis=1)
election_2019_age
```

```python
election_2019_age = election_2019_age.drop(['over65_m', 'over65_f'], axis=1)
election_2019_age
```

## Try to group by sex

```python
year_2019 = pd.melt(election_2019_age, id_vars=['Municipality', 'Province', 'Region'], \
                   var_name='Age').sort_values('Province')
year_2019
```

```python
year_2019['gender'] = year_2019.Age.str.split(pat="_", expand=True)[1]  
year_2019
```

## Clean the columns

Lower them rearrange them

```python
year_2019.columns = year_2019.columns.str.lower()
year_2019
```

```python
year_2019 = year_2019[['region', 'province', 'municipality', 'gender', 'age', 'value']]
year_2019
```

```python
year_2019 = year_2019.reset_index(drop=True)
year_2019
```

```python
year_2019.age = year_2019.age.str.replace(r'_f', '_y', regex=True)
year_2019.age = year_2019.age.str.replace(r'_m', '_y', regex=True)
year_2019
```

## Make initial analysis

```python
year_2019.groupby('age').value.sum().sort_values(ascending=False)
```

## Group the years according to Generation

```python
year_2019.age = year_2019.age.str.replace("_y", "", regex=False)
year_2019.age = year_2019.age.str.replace("over", "", regex=False)
year_2019
```

```python
year_2019.age = year_2019.age.astype(int)
year_2019
```

```python
# Define conditions for each generation
conditions = [
    (year_2019['age'] >= 79),        # Silent Generation
    (year_2019['age'] >= 60),        # Baby Boomers
    (year_2019['age'] >= 44),        # Generation X
    (year_2019['age'] >= 28),        # Millennials (Gen Y)
    (year_2019['age'] >= 12),        # Generation Z
]

# Define corresponding generation labels
choices = [
    'Silent Generation',
    'Baby Boomers/Silent Generation',
    'Generation X',
    'Millennials (Gen Y)',
    'Generation Z',
]

# Create the 'generation' column based on the conditions
year_2019['generation'] = np.select(conditions, choices, default='Unknown')

# Show the result
year_2019
```

## Fix the over 65 columns

Those people over 65 can be part of the Silent Generation of Baby Boomers. We don't have a breakdown of those.

**NOTE: For 2022, this is over 60 years old**

```python
year_2019.age = year_2019.age.astype(str)
year_2019
```

```python
year_2019.age = year_2019.age.str.replace('60', '60_and_over', regex=False)
year_2019
```
