```python
import pandas as pd
import json
import numpy as np
import geopandas as gpd
import fuzzy_pandas as fpd
from plotnine import *
pd.set_option('display.max_columns', None)
```

```python
merged =pd.read_excel("merged-data/merged_age_generation.xlsx")
merged
```

## Compute change from 2019 to 2022

```python
merged ['absolute_change'] = merged.registered_2022 - merged.registered_2019
merged ['percentage_change'] = (merged.registered_2022 - merged.registered_2019)/merged.registered_2019*100
merged
```

```python
merged.percentage_change = merged.percentage_change.round(2)
merged
```

```python
# merged.to_csv('merged_age_generation.csv', index=False)
```

## Sort by percentage change

From highest to lowest

```python
merged = merged.sort_values('percentage_change', ascending=False)
merged
```

```python
merged_province = pd.merge(province_2019, province_2022, on = ['province', 'gender', 'age', 'generation'], how='outer')
merged_province
```

```python
merged_province.percentage_change = merged_province.percentage_change.round(2)
merged_province
```

## Province, by gender

```python
merged_province
```

```python
gender_province_2022 = merged_province.groupby(['province','gender']).registered_2022.sum().to_frame().reset_index()
gender_province_2022
```

```python
gender_province_2019 = merged_province.groupby(['province','gender']).registered_2019.sum().to_frame().reset_index()
gender_province_2019
```

```python
gender_combined = pd.merge(gender_province_2019, gender_province_2022, on = ['province', 'gender'], how='outer')

gender_combined
```

```python
gender_combined ['absolute_change'] = gender_combined.registered_2022 - gender_combined.registered_2019
gender_combined ['percentage_change'] = (gender_combined.registered_2022 - gender_combined.registered_2019)/gender_combined.registered_2019*100
gender_combined
```

```python
gender_combined = gender_combined.round(2)
gender_combined
```

```python
gender_combined.province = gender_combined.province.str.title()
gender_combined
```

```python
gender_combined.columns =['Province', 'Gender', 'Registered voters (2019)', 'Registered voters (2022)', 'Change', 'Percentage change']
gender_combined
```

```python
gender_combined.Gender = gender_combined.Gender.str.replace('m', 'MALE')
gender_combined.Gender = gender_combined.Gender.str.replace('f', 'FEMALE')
gender_combined
```

## Gender by municipality

```python
merged
```

```python
municipality_gender_2019 = merged.groupby(['municipality','province','gender']).registered_2019.sum().to_frame().reset_index()
municipality_gender_2019
```

```python
municipality_gender_2022 = merged.groupby(['municipality','province','gender']).registered_2022.sum().to_frame().reset_index()
municipality_gender_2022
```

```python
gender_combined_muni = pd.merge(municipality_gender_2019, municipality_gender_2022, on = ['municipality','province', 'gender'], how='outer')
gender_combined_muni
```

```python
gender_combined_muni ['absolute_change'] = gender_combined_muni.registered_2022 - gender_combined_muni.registered_2019
gender_combined_muni ['percentage_change'] = (gender_combined_muni.registered_2022 - gender_combined_muni.registered_2019)/gender_combined_muni.registered_2019*100
gender_combined_muni.percentage_change = gender_combined_muni.percentage_change.round(2)
gender_combined_muni
```

```python
gender_combined_muni.registered_2019 = gender_combined_muni.registered_2019.map('{:,}'.format)
gender_combined_muni.registered_2022 = gender_combined_muni.registered_2022.map('{:,}'.format)
gender_combined_muni.absolute_change = gender_combined_muni.absolute_change.map('{:,}'.format)
```

```python
gender_combined_muni.columns =['Municipality','Province', 'Gender', 'Registered voters (2019)', 'Registered voters (2022)', 'Change', 'Percentage change']
gender_combined_muni.Gender = gender_combined_muni.Gender.str.replace('m', 'MALE', regex=False)
gender_combined_muni.Gender = gender_combined_muni.Gender.str.replace('f', 'FEMALE', regex=False)
gender_combined_muni
```

```python
all_males = gender_combined_muni[gender_combined_muni.Gender=="MALE"].reset_index(drop=True)
all_males
```

```python
all_females = gender_combined_muni[gender_combined_muni.Gender=="FEMALE"].reset_index(drop=True)
all_females
```
